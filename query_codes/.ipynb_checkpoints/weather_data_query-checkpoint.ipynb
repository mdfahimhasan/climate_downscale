# Read me

This script downloads [TWC/ERA5](https://location360.assets.engineering/docs/environmental/weather/data_sources_and_variables/) weather datasets for our study region, a part of Central valley, California.


### Region of Interest (ROI)
Based on testing field site locations, we have chosen some `selected cities in Califoania` in the Central Valley, California as our Region of Interest (ROI). 

__Note:__ In this script the weather datasets are downloaded for a `buffer region around the ROI`. That's because the queried ERA5 and TWC data are of low to moderate spatial resolution (28km and 4km respectively) and in later steps of the pipeline we need to interpolate (bilinear resampling) the datasets into rasters. If we only download the data for our original ROI, the interpolation results along the borders of the ROI might not do well. So, we created a buffer for our ROI and downloaded the weather datasets for that buffer regions to maintain interpolation quality. In later stesp of the pipeline, we will mask the interpolated weather datasets with our original ROI.



### Deciding value for `query_breaks`
This script uses `query_twc_weather_by_grid` and `query_era5_weather_by_grid` functions (from `query_utils.ipynb` script) for downloading the TWC + ERA5 data. There is an argument `query_breaks` in the function. `query_breaks` decides how many chunks of hid/h3 have to be made. The functions query data for each of the chunks at a time and afterwards compile them together. Querying by chunk helps improving querying speed as it queries multiple records together. Setting the `query_breaks` value large will create many chunks and query them one by one, and it can reduce querying speed if an optimum number isn't selected. On the other hand, setting the `query_breaks` value too small will create large chunks to be queried together and it might fail. Therefore, setting an `optimum value` for the argument `query_breaks` is crucial.   

__Note__: For our current ROI buffer, adjusting the value of `query_breaks` argument isn't very crucial as the number of data points (grids) isn't much. But for querying datasets for larger areas, it might be essestial.


### File Format of queried dataset 
Here, we query and save two types of data for each of the TWC and ERA5 datasets. First is the `grid_id` shapefile which is in a GeoDataFrame (.shp) file format. Then using the `grid_id` shpaefile, the TWC/ERA5 datasets are queried and the queried dataframe is saved as a dataframe in '.parquet' format. If the data records become large, trying to save the queried dataframe as a GeoDataFrame in parquet format causes `kernel died` / `memory issue`.  Therefore, I didn't save the queried data as a GeoDataFrame. Instead, I created the `read_h3_parquet_as_geodataframe` function which can read the saved parquet file (dataframe) as geodataframe without needing to save it (there is option to save, but not utilizing it), and further processing like plotting the geodataframe, converting it to raster can be performed. I have demostrated exampled below in the script.

# Query Steps:

Downloading each of the TWC and ERA5 datasets were done with the following steps:+


__1. Querying `grid_id` covered by overall extent of the ROI Buffer:__ 
First, `grid_id` covered by the minx, miny, maxx, maxy extent of the ROI Buffer was queried and saved as a shapefile using the  TWC grids covered by minx, miny, maxx, maxx bounds of California was queried and saved as shapefile using the`query_twc_grids_for_specific_bounds()`/ `query_era5_grids_for_specific_bounds` functions.


__2. Masking `grid_id` only inside ROI Buffer:__
The queried `grid_id`s were masked with the ROI Buffer shapefile to select only the `grid_id`s inside the ROI Buffer using the`clip_grids_by_admin()` function. Further dataset query was performed only for these selected `grid_id`s. 


__3. Query and Download Weather Datasets:__
The TWC and ERA5 weather datasets were queried using the `query_twc_weather_by_grid` and ` query_era5_weather_by_grid` functions respectively. The queried datasets were saved as a dataframe (.parquet) file which consists of lat/lon and grid_id info. The dataframe can be converted to GeoDataFrame using the `read_h3_parquet_as_geodataframe` function.


__Note:__
- The function used in this script can be found in `query_utils.ipynb` in the `query_codes` folder. 
- The ROI Buffer and ROI shapefiles can be found in the `shapefiles_grids` folder in the main folder of the repo.
- Downloadind the TWC weather datasets and TWC High-Res Precipitation data follows the same steps.
- In the end of this script, ERA5 Soil Temperature L1 (0-7cm) was downloaded for a site in WIsconsin. This dataset was used for exploring any potential correlation between ERA5 Soil Temperature L1 and Satellite Land SUrface Temperature Data.

__Dataset ID (in Location360 Datasets in CSW):__

-  TWC Historical Weather data 
    - grid_id : location360-datasets.historical_weather.twc_cod_grids
    - dataset id: location360-datasets.historical_weather.twc_historical_metric_daily
- TWC High-Resolution precipitation data
    - grid id : location360-datasets.historical_weather.twc_hires_precip_grids
    - dataset id : location360-datasets.historical_weather.twc_high_resolution_precipitation_metric_daily
    
- ERA5
    - grid id : location360-datasets.historical_weather.era5_grids as era5_grids
    - dataset id : location360-datasets.historical_weather.era5_metric_daily

# Querying `TWC Historical Weather Datasets` 
The TWC weather datasets are in 4km spatial resolution.

```python
# Loading ROI Buffer and Plotting
cities_california_buffer = gpd.read_file('../shapefiles_grids/cities_California_buffer.shp')


# Plotting 
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
cities_califonia_gdf = gpd.read_file('../shapefiles_grids/selected_cities_California.shp')

cities_california_buffer.plot(ax=ax)
cities_califonia_gdf.plot(ax=ax, facecolor='none', edgecolor='black')
```

In the plot, the `black outlined` shapefile is the original ROI. The `Blue` shapefile represents the ROI Buffer. We will be querying data for the `ROI Buffer`.

### 1. Querying grid_id covered by overall extent of the ROI Buffer
Here, we are querying `grid_id` covered by the min-max extent of the ROI Buffer.

```python
# Getting min-max bound of cities of California
minx, miny, maxx, maxy = cities_california_buffer.geometry.total_bounds

# Qurying TWC grids in min-max extent
twc_extent = query_twc_grids_for_specific_bounds(minx, miny, maxx, maxy, 
                                                dataset_id='location360-datasets.historical_weather.twc_cod_grids',
                                                output_folder='../../datasets/weather_data_shapefiles/', 
                                                savename='twc_grids_cities_extent.shp')
twc_extent.plot()
```

### 2. Masking grid_id only inside ROI Buffer
The first queried `grid_id` covers the entire min-max extent of the ROI Buffer. We will mask it with ROI Buffer to extract `grid_id` only inside the ROI Buffer.

```python
# Clipping TWC grids only in the cities
twc_grids_cities = clip_grids_by_admin(grids_file=twc_extent, admin_file=cities_california_buffer, 
                                       output_folder='../../datasets/weather_data_shapefiles/', 
                                       savename='twc_grids_cities_california.shp')
twc_grids_cities.plot()
```

### 3. Query and Download Weather Datasets 

Querying the TWC weather datasets for the grid_id and saving as a dataframe in '.parquet' format.

```python
twc_weather_df_cities = query_twc_weather_by_grid(grid_shp='../../datasets/weather_data_shapefiles/twc_grids_cities_california.shp', 
                                                      dataset_id='location360-datasets.historical_weather.twc_historical_metric_daily',
                                                      date=None, date_range=['2015-06-30', '2023-05-31'],
                                                      var_columns = ['max_temperature', 'min_temperature', 
                                                                     'avg_wind_speed', 'avg_relative_humidity'],
                                                      output_folder='../../datasets/weather_data_shapefiles/', 
                                                      savename='twc_weatherdata_cities_california.parquet',
                                                      bq_project='location360-datasets', query_breaks=20,
                                                      save_as_gpd=False)
```

```python
# reading the queried data
twc_weather_df_california = pd.read_parquet('../../datasets/weather_data_shapefiles/twc_weatherdata_cities_california.parquet')
twc_weather_df_california.head()
```

__Note (File Format of queried dataset) :__ 

The downloaded TWC weather dataframe has the lat/lon info. But it's not a geodataframe. We are saving the queried dataframe as a parquet file. the `read_h3_parquet_as_geodataframe` function can be used to read the saved parquet file (dataframe) as geodataframe without needing to save it (there is option to save, but not utilizing it), and further processing like plotting the geodataframe, converting it to raster can be performed.

This step is not involved with querying directly, rather a part of data processing. In the following cell, I am giving an example for demonstration purporse

```python
# reading parquet file as geodataframe and example plotting
twc_weather_gdf_california = read_parquet_as_geodataframe(parquet_file='../../datasets/weather_data_shapefiles/twc_weatherdata_cities_california.parquet', 
                                                               grid_geometry_file='../../datasets/weather_data_shapefiles/twc_grids_cities_california.shp', 
                                                               save=False)
twc_weather_gdf_california = twc_weather_gdf_california[twc_weather_gdf_california.date == '2015-06-30'].plot(column='avg_Rhumid', legend=True)
```
