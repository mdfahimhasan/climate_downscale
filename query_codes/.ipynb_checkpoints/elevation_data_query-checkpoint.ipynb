# Read Me

This script queries data of 30m resolution [SRTM DEM](https://cswarehouse.io/datasets/location360/topography/srtm/) from the [Crop Science Warehouse](https://cswarehouse.io) (CSW).

### Region of Interest (ROI)
The data is downloaded for some `selected cities in Califoania`. This Region of Interest (ROI) is selected consisting of some parts of the Central Valley, California that has significant agriculture. The ROI selection was based on consideration that it should overlap some `Testing Fields` in that region.


### Querying for each City Individually
The SRTM DEM data is of 30m resolution, therefore, the query size is quite large. If the query was set up for the whole ROI at once, the `user query` limit for a single day from CSW exceedes. So, the query has to be set up in a way that it can be performed for individual cities. For example, to avoid exceedance of user query limit, I downloaded for a couple cities in each day. In total, it took me `5/6 days` to query for our whole ROI.


### Deciding value for `query_breaks`
This script uses `query_srtm_elevation_data_on_hid_h3` function (from `query_utils.ipynb` script) for downloading the SRTM DEM data. There is an argument `query_breaks` in the function. `query_breaks` decides how many chunks of hid/h3 have to be made. The `query_srtm_elevation_data_on_hid_h3` function queries data for each of the chunks at a time and afterwards compile them together. Querying by chunk helps improving querying speed as it queries multiple records together. Setting the `query_breaks` value large will create many chunks and query them one by one, and it can reduce querying speed if an optimum number isn't selected. On the other hand, setting the `query_breaks` value too small will create large chunks to be queried together and it might fail. Therefore, setting an `optimum value` for the argument `query_breaks` is crucial.   


### File Format of queried dataset 
The downloaded SRTM DEM dataframe has the hid + h3 info. But it's not a geodataframe. We are saving the queried dataframe as a parquet file. And trying to save the dataframe as a geodataframe in parquet format causes `kernel died` / `memory issue`.  Therefore, I didn't save it as a geodataframe. Instead, I created the `read_h3_parquet_as_geodataframe` function which can read the saved parquet file (dataframe) as geodataframe without needing to save it (there is option to save, but not utilizing it), and further processing like plotting the geodataframe, converting it to raster can be performed. I have demostrated an example below in the script.

## Loading Region of interest (ROI)

A part of `central valley, California` was selected as the ROI.

```python
cities_califonia_gdf = gpd.read_file('../shapefiles_grids/selected_cities_California.shp')
cities_califonia_gdf.plot()
```

```python
cities_califonia_gdf.head()
```

## Downloading Data for selected Cities of California

There are 30 cities in our selected ROI. The city `Coalinga` is very large generating large query size. So, I spliitted that into two parts `Coalinga CCD` and `Coalinga CCD 2`. This makes the total number of polygons `31` in the `california_cities` dataframe. 

We will load the shapefile as geoDataFrame (gdf) and loop through each city in the gdf. The `query_srtm_elevation_data_on_hid_h3` function will query data for each of the cities individually and save the queried data as a parquet file individually with the city's name in the suffix. 

```python
california_cities = cities_califonia_gdf['city'].to_list()
print(len(california_cities))
california_cities
```

```python
# Looping through each city
for city in california_cities:
    print(f'Downloading Elevation data for {city} city...')
    city_gdf=cities_califonia_gdf[cities_califonia_gdf['city']==city]

    # generating h3_hid record for the city
    h3_hid_name = f'h3_hid_{city}_city.shp'
    h3_hid_gdf = make_h3_hid_using_geodataframe(geom_df=city_gdf, h3_level=11, polygon_col='geometry', 
                                                output_folder='../../datasets/elevation_data_shapefiles/', 
                                                savename=h3_hid_name)
    # Qurying data for individual city and saving them
    elevation_data_name = f'elevation_{city}_city.parquet'
    query_srtm_elevation_data_on_hid_h3(hid_h3_df=h3_hid_gdf, output_folder='../../datasets/elevation_data_shapefiles/', 
                                        savename=elevation_data_name, query_breaks=300, hid_col='hids', h3_col='h3')
```
