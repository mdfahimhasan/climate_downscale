# Read me

This script processes downloaded soil250 V1 and V2 data (https://cswarehouse.io/datasets/location360/soil/soil250/) that have been downloaded from CSW. The query codes of weather data  are in `/data-pipeline-env-model/query_codes/weather_data_query.ipnyb'`. 

The soil250 V1 and V2 datasets consists of the following variables-

__soil250 V1__
  1. Saturated water concent (awct) 
  2. Available water capacity (wwp)
  
__soil250 V2__

  3. Bulk density
  4. clay percentage
  5. Silt percentage
  6. Sand percentage
  7. Organic Matter (om)
  8. pH 
  9. Cation exchange capacity (cec)
 10. Average  Nitrogen
 11. Organic carbon content (soc)

# Data Processing Steps

The following steps have been followed for processing soil250 datasets. Both `soil250 V1 and V2` datasets have been processed through similar procedures.

__1. Compiling data for individual cities into a single parquet/dataframe:__ The soil250 V1 and V2 datasets have `250m spatial resolution`. Due to large query size for this high spatial resolution, querying soil250 datasets for the whole Region of Interest (ROI) at once wasn't feasible.  Instead, data was downloaded for individual cities (see `../query_codes/soil250_data_query.ipynb` for detail steps). These individual queried files for each city ('.parquet' files) were compiled to a single dataframe ('.parquet' format) file using `compile_soil250_data_in_single_dataframe()` function.

__2. Rasterizing attributes:__ Though the native resolution of soil250 datasets are in `250m`, the attibutes of the soil250 dataset were rasterized to `100m resolution` using `process_soil250_data()` function. This function rasterizes the data and fills gaps (nodata) to make continuous rasters. Instead of rasterizing to native `250m` resolution, we chose `100m` as the target resolution because 100m is the resolution of the satelllite datasets and we will later compile all datasets to a `100m version of Analytical-Ready Dataset (ARD)`. 
     
   - __Nodata Gap Filling:__ The original soil250 datasets have some `nodata gaps` which overlaps some of the testing sites. TO create continuous rasters, I filled these gaps by interpolation. The `process_soil250_data()` function handles this interpolation with `fill_nodata_by_interpolation()` function inside it. The `fill_nodata_by_interpolation()` function uses [rasterio fillnodata](`fill_nodata_by_interpolation()`) for the interpolation. I have chosen `60 nearby pixels` to use in interpolation based on some trial and error.




### Note
- __Region of Interest (ROI):__ We chose some `selected cities in Califoania` as the Region of Interest (ROI). The ROI consists of some parts of the Central Valley, California that has significant agriculture. The ROI selection was based on consideration that it should overlap some `Testing Fields` in that region.
- We also created a `4km ARD` for which the soil250 datasets needed to be rasterized/resampled to 4km resolution. This 4km rasterization/resampling step was performed in the `../ARD_compilation/soil_elevation_ARD_4km.ipynb` script.
- The functions used in this script can be found in `process_utils.ipnyb` script.
- The soil250 datasets have been queried from Crop Science Warehouse using codes/scripts that can be found in `../query_codes` folder.
- __Reference raster::__ `cities_California_100m_refraster` 100m raster has been used as the reference raster in this script. Refer to `Discussion.ipynb` for detail about creation of reference raster.

# V1. Processing `Soil250 V1` data for Cities of California

## 1. Compiling data for individual cities into a single parquet/dataframe

Soil250 datasets were queried for individual cities and each city has a separate parquet file. In this step, we are merging/compiling them into 01 dataframe/parquet file.

```python
soil250_data_dir = '../../datasets/soil_data_shapefiles'
output_parquet = '../../datasets/processed_parquet_data/cities_California/soil250_V1.parquet'

compiled_gdf = compile_soil250_data_in_single_dataframe(input_data_dir=soil250_data_dir, 
                                                     output_parquet=output_parquet, search_by='*V1*.parquet')
```

```python
compiled_gdf.head()
```

```python
compiled_gdf['awct_0cm'] = compiled_gdf['awct_0cm'].astype('float')
compiled_gdf.plot(column='awct_0cm')
```

## 2. Rasterizing attributes

The soil250 datasets are in `native 250` resolution. But as we will be needing 100m dataset for 100m ARD, here I am rasterizing all the attributes to 100m resolution. This step is followed by an additional `nodata gap filling` step. Refer to detail discussion in the beginning of this script.

__Note-__ For the `4km ARD` rasterization/resampling step was performed in the `../ARD_compilation/soil_elevation_ARD_4km.ipynb` script.

```python
soil250_parquet = '../../datasets/processed_parquet_data/cities_California/soil250_V1.parquet'
exclude_cols = ['h3', 'hid', 'hids', 'geometry',
                'awct_0cm', 'awct_5cm', 'awct_15cm', 'awct_30cm',
                'wwp_0cm', 'wwp_5cm', 'wwp_15cm', 'wwp_30cm']

process_soil250_data(soil250_file=soil250_parquet, ref_raster=cities_California_100m_refraster, 
                     output_dir='../../datasets/soil250_raster_data/cities_california/soil250_data',
                     remove_cols=exclude_cols, interpolation_distance=60)
```

```python
awct_data = '../../datasets/soil250_raster_data/cities_california/soil250_data/average_awct/average_awct.tif'
img_arr = rio.open(awct_data).read(1)

show(img_arr, cmap='RdYlBu', vmin=36, vmax=42)
```
