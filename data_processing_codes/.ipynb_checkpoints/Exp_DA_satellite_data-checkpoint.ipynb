# 1. Investigating Satellite Datasets

The satellite datasets consists of LST and soil moisture data. Here, we investigate the data to see to see which sites the datasets belong to, is there any overallping, etc.

```python
# Loading Woodland BBox
woodland_bbox = '../shapefiles_grids/Woodland_bbox.shp'
woodland_bbox = gpd.read_file(woodland_bbox)
print(woodland_bbox.bounds)
woodland_bbox.plot()
```

```python
# Going through each date for LST data and identifying if the datasets are only limited to Woodland, CA and WI site or more

folder = '../../datasets/satellite_raster_data/All'
temperature_datasets = glob(os.path.join(folder, '*temperature*.tif'))


# Initiating some lists for date-wise tracking
dates = []
num_total_datasets = []
num_unique_datasets= []
unique_sites_id = []

# Looping through each unique date
for date in unique_dates:
    # collecting all temperature datasets in a list 
    filtered_datasets = [i for i in temperature_datasets if date in i]
    num_datasets = len(filtered_datasets) # number of total temperature datasets in a date

    
    bbox_list = []  # empty list for adding all bounding box tuples in a date
    site_number_list = []  # empty list for adding all site numbers in a date

    # reading filtered temperature datasets for a date and saving bounding extent is a list
    for i in filtered_datasets:
        site_number = i.rsplit('_')[-2]
        site_number_list.append(site_number)
        
        bounds = rio.open(i).bounds
        minx, miny, maxx, maxy = bounds
        bbox_tup = tuple((minx, miny, maxx, maxy))
        bbox_list.append(bbox_tup)
    
    # Calculating number of unique bounding boxes (dataset) in a date
    unique_bbox = list(set(bbox_list))
    num_unique_bbox = len(unique_bbox)
    
    # Extracting site number info for each unique bounding boxes (dataset)
    site_index_for_unq_bbox = [bbox_list.index(bbox) for bbox in unique_bbox]
    selected_site_numbers = [site_number_list[i] for i in site_index_for_unq_bbox]

    # Appending info to date-wise tracking lists
    dates.append(date)
    num_total_datasets.append(num_datasets)
    num_unique_datasets.append(num_unique_bbox)
    unique_sites_id.append(selected_site_numbers)
    
# adding datewise information to a dataframe    
LST_unique_df = pd.DataFrame()
LST_unique_df['date'] = dates
LST_unique_df['total_datasets'] = num_total_datasets
LST_unique_df['unique_datasets'] = num_unique_datasets
LST_unique_df['unique_sites'] = unique_sites_id
```

### Date 2017-07-05 (5 unique sites)

```python
# Selecting unique sites with LST data for a date
date = '2017-07-05'
sites = [832, 874, 873, 877, 1151]

folder = '../../datasets/satellite_raster_data/All'
datasets = glob(os.path.join(folder, f'*{date}*.tif'))  # filtering for dates
datasets = [i for i in datasets if 'temperature' in i]  # filtering for temperature

filtered_datasets = []

for data in datasets:
    for s in sites:
        if str(s) in data:
            filtered_datasets.append(data)    
            
filtered_datasets                          
```

```python
# Loading Woodland Bbox data
woodland_bbox = '../shapefiles_grids/Woodland_bbox.shp'
woodland_bbox = gpd.read_file(woodland_bbox)
wd_minx, wd_miny, wd_maxx, wd_maxy = woodland_bbox.total_bounds
woodland_poly = Polygon([[wd_minx, wd_miny], [wd_minx, wd_maxy], [wd_maxx, wd_maxy], [wd_maxx, wd_miny], [wd_minx, wd_miny]])


# Plotting
num_cols = 3
num_rows = 2
filtered_images = filtered_datasets

fig, ax = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(8, 6))

for image in filtered_images:
    date = data.rsplit('_')[-1].split('.')[0]
    site =  image.rsplit('_')[-2]
    
    img = rio.open(image)
    minx, miny, maxx, maxy = img.bounds
    img_poly = Polygon([[minx, miny], [minx, maxy], [maxx, maxy], [maxx, miny], [minx, miny]])
    
    row = filtered_images.index(image) // num_cols
    col = filtered_images.index(image) % num_cols
    
    if img_poly.intersects(woodland_poly): # checking intersection with Woodland bbox and plotting them separately
        show(img, ax=ax[row, col], cmap='Spectral_r')
        woodland_bbox.plot(facecolor='none', edgecolor='green', ax=ax[row, col])
        ax[row, col].set_title(f'Date: {date}, Site: {site}, Woodland', fontsize=9)
    else:
        show(img, ax=ax[row, col], cmap='Spectral_r')
        ax[row, col].set_title(f'Date: {date}, Site: {site}', fontsize=9)

plt.tight_layout()
```

# 3. Calculating how many data/dates available for Woodland Site

We found three (03) unique site locations `Woodland site, CA`, `Wisconsin site`, and `Mexico Site` with LST data from previuos analysis. As our ROI for the `100m ARD` is Woodland site, here we how many days of data are available for Woodland site.

### Woodland site

```python
# Woodland min-max extent
reference_LST_raster =  '../../datasets/satellite_raster_data/All/land_surface_temperature_Veg_US_Crop_876_2018-08-06.tif'
ref_minx, ref_miny, ref_maxx, ref_maxy= rio.open(reference_LST_raster).bounds
ref_minx, ref_miny, ref_maxx, ref_maxy
```

```python
# Listing unique dates for which datasets are available
dates = []

for data in all_datasets:
    date = data.rsplit('_')[-1].split('.')[0]
    dates.append(date)

unique_dates = np.unique(dates)
print('Unique dates:', unique_dates)
```

```python
# Going through each date for LST data and selecting only temperature datasets
folder = '../../datasets/satellite_raster_data/All'
temperature_datasets = glob(os.path.join(folder, '*temperature*.tif'))

# empty list to store relevant info
date_list = []
site_number_list =[]  # store site number
data_filepath_list = []  # stores data path

# Looping through each unique date and selecting only a single dataset (will not consider the overlapping datasets) that covers whole Woodland site
for date in unique_dates:
    # collecting all temperature datasets in a list 
    filtered_datasets = [i for i in temperature_datasets if date in i]
        
    # selecting a single data for Woodland site and not considering the overlapping datasets
    for each in filtered_datasets:
        minx, miny, maxx, maxy= rio.open(each).bounds
        if (maxx==ref_maxx) & (maxy==ref_maxy) & (minx==ref_minx) & (miny==ref_miny): 
            woodland_whole_data = each
            
            # site number extraction
            site_number = os.path.basename(woodland_whole_data).rsplit('_')[-2]

            # appending info to date_list, site_number_list, data_filepath_list
            date_list.append(date)
            site_number_list.append(site_number)
            data_filepath_list.append(woodland_whole_data)
            break

woodland_df = pd.DataFrame({'date': date_list, 'site_number': site_number_list, 'data_filepath': data_filepath_list})
makedirs(['EDA_satellite_results'])

woodland_df.to_csv('./EDA_satellite_results/Woodland_LST_date_count.csv', index=False)
```

# 4. Looking for correlation between ERA5 Surface Temperature and satellite LST data for Woodland

The ERA5 dataset has `soil temperature L1 (0-7cm)` dataset. In this section, we try to compare between ERA5 soil temperature and LST at daily scale for the `Woodland site` to see if there is any correlation between them.

### 4.1 Converting multi band LST data to single band

The satellite LST datasets have two (02) bands. But only `band 1` consists of the actual data. At this step, we convert the satellite LST data to single band. We also scale the data and convert the unit from kelvin to deg C. For detail processing steps, look into `processing_daily_LST_data.ipynb` script.

```python
reference_LST_raster =  '../../datasets/satellite_raster_data/All/land_surface_temperature_Veg_US_Crop_876_2018-08-06.tif'
ref_minx, ref_miny, ref_maxx, ref_maxy= rio.open(reference_LST_raster).bounds

# creating a polygon geodataframe/shapefile using reference raster bounding box
# will use this to mask ERA5 datasets
ref_geom = box(ref_minx, ref_miny, ref_maxx, ref_maxy)
ref_gdf = gpd.GeoDataFrame({'id': 1, 'geometry': [ref_geom]})
ref_gdf = ref_gdf.set_crs('EPSG: 4326')
print(ref_gdf.crs)
ref_gdf.plot()
ref_gdf.to_file('../shapefiles_grids/LST_Woodland_bbox.shp')
```

```python
# Listing unique dates for which datasets are available
dates = []

for data in all_datasets:
    date = data.rsplit('_')[-1].split('.')[0]
    dates.append(date)

unique_dates = np.unique(dates)
print('Unique dates:', unique_dates)
```

```python
# Going through each date for LST data and selecting only temperature datasets
folder = '../../datasets/satellite_raster_data/All'
temperature_datasets = glob(os.path.join(folder, '*temperature*.tif'))

# output_dir
single_band_output_dir = '../../datasets/satellite_raster_data/Woodland/LST_single_band'

# Looping through each unique date and selecting only a single dataset (will not consider the overlapping datasets) that covers whole Woodland site
for date in unique_dates:
    # collecting all temperature datasets in a list 
    filtered_datasets = [i for i in temperature_datasets if date in i]
        
    # selecting a single data for Woodland site and not considering the overlapping datasets
    for each in filtered_datasets:
        minx, miny, maxx, maxy= rio.open(each).bounds
        if (maxx==ref_maxx) & (maxy==ref_maxy) & (minx==ref_minx) & (miny==ref_miny): 
            woodland_whole_data = each
            
    # saving raster as single band        
    save_raster_as_single_band(input_raster=woodland_whole_data, 
                               output_dir=single_band_output_dir, 
                               change_dtype=True, scale=True, scale_factor=0.01,
                               subtract_value=True, value=273.15, 
                               reformat_date=True, nodata=-9999)
        
```
